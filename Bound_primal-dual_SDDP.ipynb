# Comparing convergence of Primal and Dual SDDP

Reproducing results of Leclère et al. "Exact converging bounds for Stochastic Dual Dynamic Programming via Fenchel duality." (2018).

```julia
# we use Matplotlib for plots
using PyPlot
# import source code
include("src/dualSDDP.jl"); # main code
include("src/dp.jl"); # to compute primal inner approx "à la Philpott"
include("src/innerapprox.jl") # to compute dual inner approx
include("src/upperbound.jl") # to simulate Monte Carlo upper bounds
```

```julia
# Defining problem and algorithm parameters
Horizon = 12 # Or 36 -- Time horizon
Support = 10 # discretization of noise per time step

IterationNumber = 1000 # number of iteration of SDDP
NMC = 10000 # Number of Monte-Carlo simulation
IterationMC = 50 # Compute Monte-Carlo evaluation every IterationMC iterations
```

```julia
# define a production transport problem MPTS
mpts = MPTS([:FRA, :GER, :ESP, :UK, :PT, :ITA, :SUI, :BEL], Horizon, Support);
```

```julia
# Run a primal-dual SDDP algorithm
#include("src/dualSDDP.jl");
srand(1111)
psddp = initprimal(mpts);
dsddp = initdual(mpts, psddp);

ubd, timedual, ubp_mc, stdp, trajp = runjoint!(psddp, dsddp,
    nbsimu=NMC, maxiterations=IterationNumber, Δsimu=IterationNumber*2 ); #run primal-dual, returning dual upperbound
lbp = psddp.stats.lower_bounds; # getting primal upperbounds

writecsv("res/res_ubd_$(Horizon).csv", ubd)
writecsv("res/res_lbp_$(Horizon).csv", lbp)

```

```julia
#include("src/innerapprox.jl")
#include("src/upperbound.jl")
#compute Monte Carlo bounds (on the same scenarios)

c_oa,std_oa = computeprimalMC(psddp, IterationMC, IterationNumber, IterationMC, nscen=NMC);
writecsv("res/res_coa_$(Horizon)_$(IterationMC).csv", c_oa)
writecsv("res/res_stdoa_$(Horizon)_$(IterationMC).csv", std_oa)
c_ia,std_ia = computedualMC(psddp,dsddp, IterationMC, IterationNumber, IterationMC, nscen=NMC);
writecsv("res/res_cia_$(Horizon)_$(IterationMC).csv", c_ia)
writecsv("res/res_stdia_$(Horizon)_$(IterationMC).csv", std_ia)
```

```julia
#include("src/dp.jl")
# Compute inner bound à la "Philpott et al." with additional regularisation

#srand(1111)
# run primal sddp to get points
#psddp2 = initprimal(mpts)
#println("running Primal SDDP")
#_, _,  trajs = runprimal!(psddp2, maxiterations=IterationNumber, Δsimu=IterationNumber +1)


ub_philpott = Float64[]
# compute inner bound 
println("computing inner bound")
for it in IterationMC:IterationMC:IterationNumber
    ub = solvedp!(psddp.spmodel, trajp[1:it])[1,1]
    println("Iteration ", it, " ub-p = ", ub)
    push!(ub_philpott,ub)
end
writecsv("res/res_ubphilpott_$(Horizon).csv", ub_philpott)
```

```julia
# reading saved files
Horizon = 36 

ubd = readcsv("res/res_ubd_$(Horizon).csv")
lbp = readcsv("res/res_lbp_$(Horizon).csv")

c_oa = readcsv("res/res_coa_$(Horizon)_$(IterationMC).csv")
std_oa = readcsv("res/res_stdoa_$(Horizon)_$(IterationMC).csv")
c_ia = readcsv("res/res_cia_$(Horizon)_$(IterationMC).csv")
std_ia = readcsv("res/res_stdia_$(Horizon)_$(IterationMC).csv")

ub_philpott = readcsv("res/res_ubphilpott_$(Horizon).csv")
```

```julia
figure(figsize=(9,6))

MC_points = [i for i in 1:IterationNumber if i%IterationMC==0]

plot(ubd, c="darkblue", lw=4, label="UB-D")
plot(lbp, c="darkred", lw=4, label="LB-P")
#plot(MC_points,c_oa+2*std_oa/1000, c="red", lw=1,ls="--",marker="+", mew="5", label="MC UB OA")
plot(MC_points,c_oa, c="red", lw=2,ls="--",marker="+", mew="5", label="MC OA")
plot(MC_points,c_ia, c="blue", lw=2,ls="-.",marker=".", mew="5", label="MC IA")
plot(MC_points,ub_philpott, c="lightblue", lw=2,ls=":",marker="1", mew="5", label="UB-P")
legend(loc=4)
xlabel("Iterations")
            ylim(.95lbp[end], 1.05lbp[end])
```
